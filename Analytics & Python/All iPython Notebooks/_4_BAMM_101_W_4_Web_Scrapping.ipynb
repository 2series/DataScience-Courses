```python
import requests
from bs4 import BeautifulSoup
```

### The http Request-Respons cycle

```python
url = "https://www.simplyrecipes.com/recipes/type/baking/"
response = requests.get(url,stream = False)
if response.status_code == 200:
    print("Success")
else:
    print("Failure",response.status_code)
```

```python
searching = input("Please enter the things you want to see in a recipe: ")
url = "https://www.simplyrecipes.com/?s=" + searching
```

```python
response = requests.get(url)
if response.status_code == 200:
    print("Success")
else:
    print("Failure",response.status_code)
```

### Set up a BeautifulSoup object

```python
results_page = BeautifulSoup(response.content,'lxml')
print(results_page.prettify())
```

### BS4 functions
#### find_all() 
Finds all instances of a specified tag and returns a __ResultSet__ (a list)

```python
all_a_tags = results_page.find_all('a')
print(type(all_a_tags))
print(all_a_tags)
```

__find()__ finds the first instance of a specified tag and returns a bs4 element

```python
div_tag = results_page.find('div')
print(div_tag.prettify())
```

BS4 functions can be recursively applied on elements

```python
div_tag.find('a')
```

```python
div_tag.find('a').find('span')
```

Both __find__ as well as __find_all__ can be qualified be css selectors

- using selector=value
- using a dictionary

```python
results_page.find_all('h2',class_='entry-title')
```

```python
results_page.find('h2',{'class':'entry-title'})
```

__get_text()__ returns the marked up text (the content)enclosed in a tag
- returns a string

```python
results_page.find('h2',{'class':'entry-title'}).get_text()
```

__get()__ returns the __value__ of a tag attribute and returns a string

```python
entry_title_tag = results_page.find('ul',{'class':'entry-list'}).find('a').get('href')
entry_title_tag
```
