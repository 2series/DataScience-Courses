```python
import math
import statistics

def describe(key):
    lst = []
    for i in range(len(scripts)):
        lst.append(scripts[i][key])

    n = len(lst)
    total = sum(lst)
    avg = total/n
    s = math.sqrt(sum([(i-avg)**2 for i in lst])/n)
    ls = sorted(lst)



    med = statistics.median(lst)
    
    lq = ls[:len(lst)//2]
    uq = ls[len(lst)//2:]
    


    q25 = statistics.median(lq)
    q75 = statistics.median(uq)



    return (total, avg, s, q25, med, q75)
```
